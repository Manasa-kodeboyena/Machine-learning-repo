```python
import pandas as pd
data = {
    'rollno' : [101,102,103,104,105],
    'Name' : ['Ravi','Priya','kiran','Anu','sita'],
    'Marks' : [85,92,78,88,95]
}
df = pd.DataFrame(data)
display(df)
print(df.head(3))
avg_marks = df['Marks'].mean()
print(avg_marks)
highest_marks_name = df.loc[df['Marks'].idxmax(), 'Name']
lowest_marks_name = df.loc[df['Marks'].idxmin(), 'Name']
print(highest_marks_name)
print(lowest_marks_name)
conditions = [
    (df['Marks'] > 90),
    ((df['Marks'] >=80) & (df['Marks']<=89)),
    (df['marks'] < 80)
]
choices ['A','B', 'C']
df['Grade'] = np.select(conditions, choices)
print(df)
```

```python
import pandas as pd
data = {
    'Product':['Laptop','Mouse','Keyboard','Monitor'],
    'Price' : [50000,500,1200,10000],
    'Quantity':[10,50,30,15]
}
df = pd.DataFrame(data)
display(df)
#Calculate and add the total value of each product
df['Total Value'] = df['Price'] *df['Quantity']
display(df)
#highest total value product
highest_total_value_product = df.loc[df['Total Value'].idxmax(), 'Product']
print(highest_total_value_product)
#calculate the overall inventory value
total_value = df['Total Value'].sum()
print(total_value)

```
