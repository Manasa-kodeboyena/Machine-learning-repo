```python
import pandas as pd

df = pd.read_csv(r"C:\Users\kodeb\Downloads\electronics..csv")

# Remove RID column if present
if "RID" in df.columns:
    df = df.drop("RID", axis=1)

target = df.columns[-1]
total = len(df)
print("Total records :", total)
class_counts = df[target].value_counts()
print("Class counts:")
print(class_counts) 
#-------------prior probablility-------------
prior = {}
for cls in class_counts.index:
    prior[cls] = class_counts[cls]/total
print("Prior probabilities:")
for cls in prior:
    print(f"{cls}: {prior[cls] :.4f}")
yes_df = df[df[target] == 'yes']
no_df = df[df[target] == 'no']

#----------------test-----------------------
test = {}
for feature in df.columns[:-1]:
    test[feature] = input(f"Enter feature name {feature}").lower()
    
print("-----------Likelihoods for YES : -----------")
yes_prob =prior['yes']
for feature, value in test.items():
    count = len(yes_df[yes_df[feature]== value])
    likelihood = count/len(yes_df)
    yes_prob *= likelihood
    

print("-----------Likelihoods for NO : -----------")
no_prob = prior['no']
for feature, value in test.items():
    count = len(no_df[no_df[feature] == value])
    likelihood = count/len(no_df)
    no_prob *= likelihood
print(f"Probability of YES: {yes_prob:.4f}")
print(f"Probability of NO: {no_prob:.4f}")
#-----------Final Prediction----------------
if yes_prob > no_prob:
    print("The predicted class is: YES")
else:
    print("The predicted class is: NO")
```
